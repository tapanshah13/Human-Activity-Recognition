```python
# Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Create Classifier
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# Open Image
image = cv2.imread('C:/Users/Kush/Pictures/Kush/40I90009.jpg')
fix_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# Detect Faces
faces = face_classifier.detectMultiScale(image,1.3,5)
```

```python
# No face detected
if faces is ():
    print('No face found')
```

```python
# It's kind of magic!
def detect_face(fix_img):
    face_rectangle = face_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in face_rectangle:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w,y+h),
                     (255,0,0),
                     4)
        
        return fix_img
```

```python
result = detect_face(fix_img)
plt.imshow(result)
```

# Face & Eyes Detection

```python
# create classifiers
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
# Open Image
img = cv2.imread('C:/Users/Kush/Pictures/Family Photo/Family.jpg')
```

```python
# fix image
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
# Face classifier
faces = face_classifier.detectMultiScale(fix_img,1.35,15)
```

```python
# No face detected 
if faces is ():
    print('No Faces Found')
```

```python
# It's kind of magic
def detect_faces_eyes(fix_img):
    
    face_rectangle = face_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in face_rectangle:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w,y+h),
                     (255,0,0),
                      7)
        eyes_rectangle = eye_classifier.detectMultiScale(fix_img)
        for(ix,iy,iw,ih) in eyes_rectangle:
            cv2.rectangle(fix_img,
                         (ix,iy),
                         (ix+iw,iy+ih),
                         (0,0,255),
                         5)
            
        return fix_img
```

```python
result = detect_faces_eyes(fix_img)
plt.imshow(result)
```

# Pedestrians Detection

```python
# create classifier 
body_classifier = cv2.CascadeClassifier('haarcascade_fullbody.xml')
```

```python
# Capture video
cap = cv2.VideoCapture('StoreDenseFlow.mp4')
```

```python
# While loop
while cap.isOpened():
    
    # Read the capture
    ret , frame = cap.read()
    
    # Pass the frame to the Classifier
    bodies = body_classifier.detectMultiScale(frame,1.2,5)
    
    # if statement
    if ret == True:
        
        # BOund Boxes to Identified Bodies
        for (x,y,w,h) in bodies:
            cv2.rectangle(frame,
                         (x,y),
                         (x+w,y+h),
                         (25,125,255),
                         5)
            cv2.imshow('Pedestrains',frame)
            
            
        # Exit with Esc button
        if cv2.waitKey(1) & 0xFF ==27:
            break
            
        # else statement
    else:
        break
            
# release the capture & destroy all windows
cap.release()
cv2.destroyAllWindows()
```

# Car Detection

```python
# create classifier
car_classifier = cv2.CascadeClassifier('haarcascade_car.xml')
```

```python
# Capture video
cap = cv2.VideoCapture('Vehicles.mp4')
```

```python
# Loop once video is successfully loaded
#cap = cv2.VideoCapture(0)

#Check whether user selected camera is opened successfully.

if not (cap.isOpened()):
    print("Could not open video device")

while cap.isOpened():
    # Read first frame
    ret, frame = cap.read()
    # gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
   
    # Pass frame to our car classifier
    cars = car_classifier.detectMultiScale(frame, 1.4, 2)
    
    # Extract bounding boxes for any bodies identified
    for (x,y,w,h) in cars:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        cv2.imshow('Car Detection', frame)

    if cv2.waitKey(1) == 27: # 27 is the Esc button
        break

cap.release()
cv2.destroyAllWindows()
```

# Numer plate detection


```python
img = cv2.imread('car_plate.jpg')
```

```python
# Convert Image
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
# display it
display(img)
```

```python
# create classifier
plate_classifier = cv2.CascadeClassifier('C:/Users/Kush/Documents/Computer Vision/haarcascades/haarcascade_russian_plate_number.xml')
```

```python
# it's kind of magic
def detect_plate(img):
    
    plate_img = img.copy()
    
    plate_rectange = plate_classifier.detectMultiScale(plate_img,1.1,1)
    
    for(x,y,w,h) in plate_rectange:
        cv2.rectangle(plate_img,
                     (x,y),
                     (x+w,y+h),
                     (255,255,255),
                     5)
    return plate_img
```

```python
# check results
result = detect_plate(img)
display(result)
```
